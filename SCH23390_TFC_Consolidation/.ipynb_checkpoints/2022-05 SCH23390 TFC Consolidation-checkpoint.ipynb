## Experiment Info: 
- Inject Saline or SCH23390 (.05 or 0.1 mg/kg) 30 mins before training in TFC
- Train in trace fear conditioning (TFC)
    - 240s baseline
    - 20s, 3kHz, 85db tone
    - 20s trace interval
    - 3 x 0.3mA, 2s shocks
    - 240s ITI
- Tone test
    - 180s baseline
    - same tone
    - 6 trials
- 10m context test

## Training

<a id='lctr'></a>
### Load and Clean Training Data

```python
# load data
train_file = f'{path}data-raw/2021-11-10 SCH23390 TFC Train.csv'
df_train_raw = fd.load_df(train_file)

# clean data (take from MedAssociates format to simple csv table)
df_train = fd.clean_df(df_train_raw)

# add group labels to animals
df_train = fd.add_group_labels(df_train, grp_file=grp_label_file, exp_name=expt_name, grp2='Condition')


# create a df w/ values averaged for each phase (i.e. baseline, tone, etc.) w/in each animal
df_train_total = fd.total_df(df_train, grp2='Condition')

```

<a id='pltr'></a>
### Plot Training Data

```python
# binned plot (w/ time as x)

fd.tfc_plot(df_train, has_hue=True, hue='Condition', fig_name='Training')
plt.legend(fontsize = 50);

# Save figure
#plt.savefig(f'{path}/figures/{expt_name} train bins.png')
min_bins = [i for i in range(len(df_train['Component Name'].unique())) if (i+1) % 3 == 0]
min_labs = [ i+1 for i in range(len(min_bins)) ]
plt.xticks(min_bins, min_labs, rotation=0)
plt.xlabel('Time (min)', size = 36);
plt.savefig(f'{path}figs/SCH23390 TFC Train bins.png')
```

```python
# bar plot grouped by phase

fd.tfc_barplot(df_train_total, hue='Condition', order=['baseline', 'tone', 'trace'], fig_name='Training', pts=True)
plt.ylim(0, 100)

# Save figure
#plt.savefig(f'{path}/figs/SCH23390 TFC Train total.png')
```

```python
fd.tfc_plot(df_train_total, xvar = 'Phase', min_bins= False,tone_squares=False, trace_squares=False,
         shock_line=False,hue='Condition', order = ['baseline', 'tone', 'trace'], 
         errwidth=10,scale = 2, marker_size = 4500)
plt.ylim((0,100));
plt.xlabel('')
plt.ylabel('')
ax = plt.gca()
ax.legend(markerscale = .5, fontsize = 70)
ax.set_xticklabels(['Baseline', 'Tone', 'Trace'], size = 70)

#plt.savefig(f'{path}/figures/LC ArchT TFC Train Total Line.png')
```

## Tone Test

<a id='lcto'></a>
### Load and Clean Tone Test Data

```python
# load data
tone_file = f'{path}data-raw/2021-11-11 SCH23390 TFC Test.csv'
df_tone_raw = fd.load_df(tone_file)

# clean data (take from MedAssociates format to simple csv table)
df_tone = fd.clean_df(df_tone_raw)

# add group labels to animals
df_tone = fd.add_group_labels(df_tone, grp_file=grp_label_file, exp_name=expt_name, grp2= 'Condition')

# exclude animal that didn't get some shocks
#df_tone = df_tone[df_tone.Animal != '306']


# create a df w/ values averaged for each phase (i.e. baseline, tone, etc.) w/in each animal
df_tone_total = fd.total_df(df_tone, grp2='Condition')
```

<a id='plto'></a>
### Plot Tone Test Data

```python
# binned plot

fd.tfc_plot(df_tone, has_hue=True, hue='Condition', fig_name='Tone Test')
plt.legend(fontsize=50);


# Save figure
#plt.savefig(f'{path}/figures/{expt_name} tone test trial.png')
```

```python
# bar plot

fd.tfc_barplot(df_tone_total, hue='Condition', order=['baseline', 'tone', 'trace'], 
               fig_name='Tone Test', pts = False)
plt.ylim((0,80))


# Save figure
#plt.savefig(f'{path}/figures/{expt_name} tone test total bar.png')
```

```python
plt.figure(figsize = (10,10));

fd.tfc_plot(df_tone_total, xvar = 'Phase', min_bins= False,tone_squares=False, trace_squares=False,
         shock_line=False,hue='Condition', order = ['baseline', 'tone', 'trace', ], 
         errwidth=10,scale = 2, marker_size = 4500)

plt.ylim(0,79);
plt.xlabel('');
plt.ylabel('% Freezing', size = 150, weight = 'normal',labelpad = 50);

ax = plt.gca()
ax.set_xticklabels(['Baseline', 'Tone', 'Trace']);
ax.tick_params(labelsize=100);

L=ax.legend(markerscale = .5, fontsize = 100, loc = "upper left")
#L.get_texts(0).set_texts('Sal')

plt.tight_layout()
plt.savefig(f'{path}/figs/{expt_name} tone test total line.png')
```
