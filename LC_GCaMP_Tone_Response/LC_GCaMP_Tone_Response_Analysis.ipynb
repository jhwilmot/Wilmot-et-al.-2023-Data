```python
df_train_trials = pd.read_csv(
    './data-clean/LC_GCaMP_Tone_Response_Clean.csv')
```

```python
fig, ax = plt.subplots(figsize = (14,10))

fp_viz.tfc_trial_avg(df_train_trials, hue = 'dB',yvar = '465nm_dFF_norm', ax = ax, us_dur = 0, cs_dur = 0)

plt.xlim(-2,5)
plt.ylim(-2,9.9)

plt.axvline(x = 0, color = 'k', linestyle = '--', linewidth = 2)

plt.legend(fontsize = 40)
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)

plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
```

```python
event = 'tone'
t_pre = (-2,0)
t_post = (0,2)

#df_train_trials = df_train_trials.query('Trial <= 24')

df_pre = df_train_trials[df_train_trials['time_trial'].between(t_pre[0], t_pre[1])].reset_index(drop=True)
df_post = df_train_trials[df_train_trials['time_trial'].between(t_post[0], t_post[1])].reset_index(drop=True)

# add `epoch` column
df_pre['epoch'] = f'pre-{event}'
df_post['epoch'] = f'post-{event}'

# recombine values and groupby new epoch var

df_prepost = pd.concat([df_pre, df_post])

df_prepost = df_prepost.groupby(['Animal', 'time_trial', 'epoch', 'dB']).mean().reset_index()
df_prepost = df_prepost.groupby(['Animal', 'epoch','dB']).max().reset_index()
```

```python
fig, ax = plt.subplots(figsize = (14,10))

p = sns.barplot(x = 'dB', y = '465nm_dFF_norm', hue = 'epoch', hue_order = ['post-tone'],
                ci = 68, capsize=0.1, 
                data = df_prepost)

plt.ylim(0, 9.9)

plt.legend('')
ax.set_ylabel('')
plt.xlabel('')

plt.xticks(fontsize=50)
plt.yticks(fontsize=50)

plt.tight_layout()
```
