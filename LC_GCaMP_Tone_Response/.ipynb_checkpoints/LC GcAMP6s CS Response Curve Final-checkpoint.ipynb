```python
file_path = '/Users/jake/Box/WiltgenLab expts/JW LC/2021-07 LC GcAMP6s Tone Response 2nd Attempt/'
raw_data_path = f'{file_path}data-raw/'
```

```python
# import group data
df_train = fp_dat.load_session_data(file_dir=raw_data_path)
```

```python
df_train['Animal'].unique()
```

```python
df_train.loc[df_train['Animal'] == '2021-07-11 LC GcAMP6s Tone Response_1.csv', 'Animal'] = 229
df_train.loc[df_train['Animal'] == '2021-07-11 LC GcAMP6s Tone Response_2.csv', 'Animal'] = 232
df_train.loc[df_train['Animal'] == '2021-07-11 LC GcAMP6s Tone Response_3.csv', 'Animal'] = 233
df_train.loc[df_train['Animal'] == '2021-07-11 LC GcAMP6s Tone Response_4.csv', 'Animal'] = 235
```

```python


# apply 10 Hz low-pass filter
df_train_lowfilt = fp_dat.lowpass_filter(df_train, x1='465nm', x2='405nm', 
                                          cutoff=6, fsamp=100, order=4)
# resample to 10 Hz
df_train_resamp = fp_dat.resample_data(df_train_lowfilt, freq=10)

# apply linear fit for motion correction to obtain dFF
df_train_fit = fp_dat.fit_linear(df_train_resamp, Y_sig='465nm', Y_ref='405nm')

# trial level data
df_train_trials = tfc_dat.trials_df(df_train_fit, yvar='465nm_dFF', session='cs_response_2', trial_start = -10,
                                   iti_dur = 30, trace_dur = 0, cs_dur= 20, us_dur = 0)
```

```python
fp_viz.fp_trace(df_train_fit, trace = 'raw')
```

```python
dB_dict = {1:55, 2:95, 3:75, 4:85, 5:65, 
           6:65, 7:85, 8:75, 9:55, 10:95,
           11:55, 12:65, 13:85, 14:95, 15:75, 
           16:85, 17:75, 18:55, 19:95, 20:65, 
           21:95, 22:65, 23:75, 24:85, 25:55}
```

```python
df_train_trials['dB'] = df_train_trials['Trial'].map(dB_dict)
```

```python
df_train_trials['Block'] = 0

df_train_trials.loc[df_train_trials['Trial'] < 6, 'Block'] = 1
df_train_trials.loc[df_train_trials['Trial'].between(6,10), 'Block'] = 2
df_train_trials.loc[df_train_trials['Trial'].between(11,15), 'Block'] = 3
df_train_trials.loc[df_train_trials['Trial'].between(16, 20), 'Block'] = 4
df_train_trials.loc[df_train_trials['Trial'].between(21,25), 'Block'] = 5
```

```python
fig, ax = plt.subplots(figsize = (14,10))

#fp_viz.tfc_trial_avg(df_train_trials, yvar = '465nm_dFF_norm', hue = 'dB', ax = ax, us_dur = 0)
fp_viz.tfc_trial_avg(df_train_trials.query('dB == 55'), yvar = '465nm_dFF_norm', ax = ax, us_dur = 0)
plt.xlim(-5,10)
```

```python
event = 'tone'
t_pre = (-1,0)
t_post = (0,1)

#df_train_trials = df_train_trials.query('Trial <= 24')

df_pre = df_train_trials[df_train_trials['time_trial'].between(t_pre[0], t_pre[1])].reset_index(drop=True)
df_post = df_train_trials[df_train_trials['time_trial'].between(t_post[0], t_post[1])].reset_index(drop=True)

# add `epoch` column
df_pre['epoch'] = f'pre-{event}'
df_post['epoch'] = f'post-{event}'

# recombine values and groupby new epoch var

df_prepost = pd.concat([df_pre, df_post])

df_prepost = df_prepost.groupby(['Animal', 'time_trial', 'epoch', 'dB']).mean().reset_index()
df_prepost = df_prepost.groupby(['Animal', 'epoch','dB']).mean().reset_index()
```

```python
fig, ax = plt.subplots(figsize = (10,10))

p = sns.barplot(x = 'dB', y = '465nm_dFF_norm', hue = 'epoch',
                ci = 68, capsize=0.1, 
                data = df_prepost)

ax.set_ylabel(r'Mean $\Delta$F/F')

#ax.set_ylim(0,2)

plt.tight_layout()
```
