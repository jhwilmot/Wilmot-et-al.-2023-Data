```python
# import group data
df_train = fp_dat.load_session_data(file_dir=raw_data_path)
```

```python
df_train.loc[df_train['Animal'] == '2021-06-02 LC GcAMP6s Tone Response Curve_1.csv', 'Animal'] = 199
df_train.loc[df_train['Animal'] == '2021-06-02 LC GcAMP6s Tone Response Curve_2.csv', 'Animal'] = 200
df_train.loc[df_train['Animal'] == '2021-06-02 LC GcAMP6s Tone Response Curve_3.csv', 'Animal'] = 206
df_train.loc[df_train['Animal'] == '2021-06-03 LC GcAMP6s Tone Response Curve_1.csv', 'Animal'] = 209
```

```python
df_train = df_train[df_train['Animal'] != 209]

# apply 10 Hz low-pass filter
df_train_lowfilt = fp_dat.lowpass_filter(df_train, x1='465nm', x2='405nm', 
                                          cutoff=6, fsamp=100, order=4)
# resample to 10 Hz
df_train_resamp = fp_dat.resample_data(df_train_lowfilt, freq=10)

# apply linear fit for motion correction to obtain dFF
df_train_fit = fp_dat.fit_linear(df_train_resamp, Y_sig='465nm', Y_ref='405nm')

# trial level data
df_train_trials = tfc_dat.trials_df(df_train_fit, yvar='465nm_dFF', session='cs_response', trial_start = -10,
                                   iti_dur = 30, us_dur = 0)
```

```python
fp_viz.fp_trace(df_train_fit, trace = 'dFF')
```

```python
dB_dict = {1:60, 2:20, 3:70, 4:40, 5:50, 6:90, 7:30, 8:80,
          9:70, 10:40, 11:50, 12:20, 13:60, 14:90, 15:80, 16:30,
          17:70, 18:60, 19:90, 20:50, 21:20, 22:30, 23:40, 24:80,
          25:50, 26:60, 27:20, 28:40, 29:80, 30:30, 31:70, 32:90,
          33:50, 34:70, 35:90, 36:60, 37:80, 38:30, 39:20, 40:40}
```

```python
df_train_trials['dB'] = df_train_trials['Trial'].map(dB_dict)
```

```python
df_train_trials
```

```python
fig, ax = plt.subplots(figsize = (14,10))

fp_viz.tfc_trial_avg(df_train_trials.query('dB == 90'), yvar = '465nm_dFF_norm', ax = ax, us_dur = 0)
fp_viz.tfc_trial_avg(df_train_trials.query('dB == 80'), yvar = '465nm_dFF_norm', ax = ax, us_dur = 0)
fp_viz.tfc_trial_avg(df_train_trials.query('dB == 60'), yvar = '465nm_dFF_norm', ax = ax, us_dur = 0)
fp_viz.tfc_trial_avg(df_train_trials.query('dB == 20'), yvar = '465nm_dFF_norm', ax = ax, us_dur = 0)
```

```python
event = 'tone'
t_pre = (-2,0)
t_post = (0,2)

#df_train_trials = df_train_trials.query('Trial <= 24')

df_pre = df_train_trials[df_train_trials['time_trial'].between(t_pre[0], t_pre[1])].reset_index(drop=True)
df_post = df_train_trials[df_train_trials['time_trial'].between(t_post[0], t_post[1])].reset_index(drop=True)
# add `epoch` column
df_pre['epoch'] = f'pre-{event}'
df_post['epoch'] = f'post-{event}'
# recombine values and groupby new epoch var
df_prepost = pd.concat([df_pre, df_post])

df_prepost = df_prepost.groupby(['Animal', 'time_trial', 'epoch', 'dB']).mean().reset_index()
df_prepost = df_prepost.groupby(['Animal', 'epoch','dB']).max().reset_index()
```
