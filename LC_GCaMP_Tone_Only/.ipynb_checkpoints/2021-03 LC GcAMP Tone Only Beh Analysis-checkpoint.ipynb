## Training

<a id='lctr'></a>
### Load and Clean Training Data

```python
# load data
train_file = '/Users/jake/Box/WiltgenLab expts/JW LC/2020-12 LC GcAMP TFC/data-raw/beh_dat/LC GcMP TFC Train.csv'
df_train_raw = fd.load_df(train_file)

# clean data (take from MedAssociates format to simple csv table)
df_train = fd.clean_df(df_train_raw)

# add group labels to animals
df_train = fd.add_group_labels(df_train, grp_file=grp_label_file, exp_name=expt_name, grp2='Condition')


# exclude animal that didn't get some shocks
#df_train = df_train[df_train.Animal != '306']



# create a df w/ values averaged for each phase (i.e. baseline, tone, etc.) w/in each animal
df_train_total = fd.total_df(df_train, grp2='Condition')

```

<a id='pltr'></a>
### Plot Training Data

```python
# binned plot (w/ time as x)

fd.tfc_plot(df_train, has_hue=False,fig_name='Training')
plt.legend(fontsize = 50);

# Save figure
#plt.savefig(f'{path}/figures/{expt_name} train bins.png')
min_bins = [i for i in range(len(df_train['Component Name'].unique())) if (i+1) % 3 == 0]
min_labs = [ i+1 for i in range(len(min_bins)) ]
plt.xticks(min_bins, min_labs, rotation=0)
plt.xlabel('Time (min)', size = 36);
plt.savefig('/Users/jake/Box/WiltgenLab expts/JW LC/2020-12 LC GcAMP TFC/figs/LC GcAMP Train Beh binned')
```

```python
# bar plot grouped by phase

fd.tfc_barplot(df_train_total, order=['baseline', 'tone', 'trace', 'iti'], fig_name='Training',ylim = (0,100), pts=True)

# Save figure
plt.savefig('/Users/jake/Box/WiltgenLab expts/JW LC/2020-12 LC GcAMP TFC/figs/LC GcAMP Train Beh bar')
```

```python
fd.tfc_plot(df_train_total, xvar = 'Phase', min_bins= False,tone_squares=False, trace_squares=False,
         shock_line=False, order = ['baseline', 'tone', 'trace', 'iti'], 
         errwidth=10,scale = 2, marker_size = 4500)
plt.ylim((0,100));
plt.xlabel('')
plt.ylabel('% Freezing')
ax = plt.gca()
ax.legend(markerscale = .5, fontsize = 70)
ax.set_xticklabels(['Baseline', 'Tone', 'Trace', 'ITI'], size = 70)

plt.savefig('/Users/jake/Box/WiltgenLab expts/JW LC/2020-12 LC GcAMP TFC/figs/LC GcAMP Train Beh total line')
```
