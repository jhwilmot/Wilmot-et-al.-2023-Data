## Experiment Info: 
- Infuse AAV5-DIO-ChETA-eYFP into LC of TH-Cre mice
- Train in trace fear conditioning (TFC)
    - 180s baseline
    - 20s, 3kHz, 65db tone
    - 20s trace interval
    - 3 x 0.2mA, 2s shocks
    - 180s ITI
- Silence during training
    - laser on for first 2 seconds of tone, trace, shock
    - 20Hz, 5ms pulse width, ~10mW
- Tone test
    - 180s baseline
    - same tone
    - 6 trials
- 10m context test

## Training

<a id='lctr'></a>
### Load and Clean Training Data

```python
# load data
train_file = f'{path}data-raw/2021-11 LC ChR2 65dB 0.2mA TFC train.csv'
df_train_raw = fd.load_df(train_file)

# clean data (take from MedAssociates format to simple csv table)
df_train = fd.clean_df(df_train_raw)

# add group labels to animals
df_train = fd.add_group_labels(df_train, grp_file=grp_label_file, exp_name=expt_name, grp2='Condition')

# exclude animals that had loud noise during training

exclude = ['279', '280']

df_train = df_train.loc[~df_train.Animal.isin(exclude)]

# create a df w/ values averaged for each phase (i.e. baseline, tone, etc.) w/in each animal
df_train_total = fd.total_df(df_train, grp2='Condition')

```

```python
df_train_total.Animal.unique()
```

<a id='pltr'></a>
### Plot Training Data

```python
# binned plot (w/ time as x)

fd.tfc_plot(df_train, has_hue=True, hue='Condition', hue_order=['Cre-', 'Cre+'], fig_name='Training')
plt.legend(fontsize = 50);

# Save figure
#plt.savefig(f'{path}/figures/{expt_name} train bins.png')
min_bins = [i for i in range(len(df_train['Component Name'].unique())) if (i+1) % 3 == 0]
min_labs = [ i+1 for i in range(len(min_bins)) ]
plt.xticks(min_bins, min_labs, rotation=0)
plt.xlabel('Time (min)', size = 36);
plt.savefig(f'{path}/figures/LC ChR2 65dB 0.2mA TFC Train Trials.png')
```

```python
# bar plot grouped by phase

fd.tfc_barplot(df_train_total, hue='Condition', hue_order=['Cre-','Cre+'], order=['baseline', 'tone', 'trace', 'iti'], fig_name='Training', pts=True)

# Save figure
#plt.savefig(f'{path}/figures/{expt_name} train total.png')
```

```python
fd.tfc_plot(df_train_total, xvar = 'Phase', min_bins= False,tone_squares=False, trace_squares=False,
         shock_line=False,hue='Condition', order = ['baseline', 'tone', 'trace'], hue_order=['Cre-', 'Cre+'], 
         errwidth=10,scale = 2, marker_size = 4500)
plt.ylim((0,40));
plt.xlabel('')
plt.ylabel('')
ax = plt.gca()
ax.legend(markerscale = .5, fontsize = 70)
ax.set_xticklabels(['Baseline', 'Tone', 'Trace'], size = 70)

```
