```python
def DREADDs_sess_avg(df,
                     hue = None,
                     title=None,
                     yvar='465nm_dFF_znorm',
                     xvar='time',
                     fig_size=(12, 8), 
                     ax=None, **kwargs):
    """
        Plot trial-averaged dFF signal.

        Parameters
        ----------
        df : DataFrame
            Trial-level DataFrame from trials_df()
        yvar : str, optional
            Column containing fluorescence values to plot, by default '465nm_dFF_znorm'
        xvar : str, optional
            Column containing trial-level timepoints, by default 'time_trial'
        fig_size : tuple, optional
            Size of figure, by default (12, 8)
        """    
    plt.figure(figsize = (10,10)) 
    ax = plt.gca()
    
    # Used to plot each animal separately    
    if hue:
        hue_means = df.groupby([xvar, hue]).mean().reset_index()
        if hue in ['Animal', 'Trial']:         
            hue_sems = df.groupby([xvar, hue]).sem().reset_index()
        else:
            hue_sems = (df.groupby([xvar, hue, 'Animal']).mean()
                        .groupby([xvar, hue]).sem().reset_index())
        # plot the data for each hue level
        for hue_level in hue_means[hue].unique():
            x = hue_means.loc[hue_means[hue] == hue_level, xvar]
            y = hue_means.loc[hue_means[hue] == hue_level, yvar]
            yerr = hue_sems.loc[hue_sems[hue] == hue_level, yvar]
            line = ax.plot(x, y, label=f'{hue}: {hue_level}', **kwargs)
            ax.fill_between(x, y-yerr, y+yerr, facecolor=line[0].get_color(), alpha=0.15)
            ax.legend(fontsize=12)
    else:
        animal_means = df.groupby([xvar]).mean().reset_index()
        animal_stds = (df.groupby([xvar, 'Animal']).mean()
                       .groupby(xvar).std().reset_index())
        # grab variables for plotting
        x = animal_means.loc[:, xvar]
        y = animal_means.loc[:, yvar]
        yerror = animal_stds.loc[:, yvar]
        # plot the data
        line = ax.plot(x, y, **kwargs)
        ax.fill_between(x, y-yerror, y+yerror, facecolor=line[0].get_color(), alpha=0.15)
```

```python
DREADDs_sess_avg(df_grab, xvar = 'standard_time', yvar = '465nm_dFF_zscore')

plt.ylabel('dFF')
plt.title('CNO vs Saline');
```

```python

df_grab.loc[df_grab['standard_time'] < 300 , 'phase'] = 'baseline'

df_grab.loc[df_grab['standard_time'] > 300, 'phase'] = 'trial'
```

```python
norm_dat = []

for animal in df_grab['Animal'].unique():
    df_animal = df_grab.loc[df_grab.Animal == animal]
    norm_vals = []
    
    df_base = df_animal.loc[df_animal.phase == 'baseline']
    base_mean = df_base['465nm'].mean()
    base_std = df_base['465nm'].std()
    norm_vals.append((df_animal['465nm'] - base_mean)/base_std)
    norm_vals = [item for sublist in norm_vals for item in sublist]
    norm_dat.append(norm_vals)
    
norm_dat = [item for sublist in norm_dat for item in sublist]
df_grab['base_norm'] = norm_dat
    
    
    
    
    
```

```python
DREADDs_sess_avg(df_grab, xvar = 'standard_time', yvar = 'base_norm', hue = 'Group')
plt.xlim (0, 3600)
plt.ylabel('Baseline Normalized 465')

plt.axvline(x = 270, c = 'k', linewidth = 0.7, linestyle = 'dashed', alpha = 0.7 )
plt.axvline(x = 330, c = 'k', linewidth = 0.7, linestyle = 'dashed', alpha = 0.7)
plt.axvspan(270, 330, ymax = 30, color = 'k', alpha = 0.1)

plt.legend(['CNO', 'CNO + ETI (1mg)', 'CNO + ETI (2mg)','Vehicle'], prop={"size":25}, loc = 'upper right', bbox_to_anchor = (1.5,1.05))
plt.axhline(y=0, c = 'k', linestyle = 'dashed')
```
