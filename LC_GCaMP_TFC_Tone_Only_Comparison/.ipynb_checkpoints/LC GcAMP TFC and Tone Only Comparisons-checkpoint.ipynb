# Load Tone Only Data

```python
file_path = '/Users/jake/Box/WiltgenLab expts/JW LC/2021-03 LC GcAMP Tone Only/'
raw_data_path = f'{file_path}data-raw/'
train_path = f'{file_path}data-raw/train/'
d3_ctxB_hab_path = f'{file_path}data-raw/d3_ctxB_hab/'
d4_tone_path = f'{file_path}data-raw/tone test/'
d6_extinction_path = f'{file_path}data-raw/extinction/'
fig_file_path = f'{file_path}figs/'
```

```python
# import group data
df_train = fp_dat.load_session_data(file_dir=train_path)
df_tone = fp_dat.load_session_data(file_dir=d4_tone_path)
df_ext = fp_dat.load_session_data(file_dir=d6_extinction_path)
```

```python
# rename animals from file names
df_train.loc[df_train['Animal'] == '2021-03-31 LC GcAMP Tone Only Train Session_1.csv', 'Animal'] = 180
df_train.loc[df_train['Animal'] == '2021-05-18 LC GcAMP Tone Only Train_1.csv', 'Animal'] = 199
df_train.loc[df_train['Animal'] == '2021-05-18 LC GcAMP Tone Only Train_2.csv', 'Animal'] = 200
df_train.loc[df_train['Animal'] == '2021-05-18 LC GcAMP Tone Only Train_3.csv', 'Animal'] = 206
df_train.loc[df_train['Animal'] == '2021-05-18 LC GcAMP Tone Only Train_4.csv', 'Animal'] = 207
df_train.loc[df_train['Animal'] == '2021-05-18 LC GcAMP Tone Only Train_5.csv', 'Animal'] = 209

df_tone.loc[df_tone['Animal'] == '2021-04-02 LC GcAMP Tone Only Tone Test Session_1.csv', 'Animal'] = 180
df_tone.loc[df_tone['Animal'] == '2021-05-20 LC GcAMP Tone Only Tone Test_1.csv', 'Animal'] = 199
df_tone.loc[df_tone['Animal'] == '2021-05-20 LC GcAMP Tone Only Tone Test_2.csv', 'Animal'] = 200
df_tone.loc[df_tone['Animal'] == '2021-05-20 LC GcAMP Tone Only Tone Test_3.csv', 'Animal'] = 206
df_tone.loc[df_tone['Animal'] == '2021-05-20 LC GcAMP Tone Only Tone Test_4.csv', 'Animal'] = 209

df_ext.loc[df_ext['Animal'] == '2021-04-05 LC GcAMP Tone Only Extinction Session_1.csv', 'Animal'] = 180
df_ext.loc[df_ext['Animal'] == '2021-05-21 LC GcAMP Tone Only Extinction_1.csv', 'Animal'] = 199
df_ext.loc[df_ext['Animal'] == '2021-05-21 LC GcAMP Tone Only Extinction_2.csv', 'Animal'] = 200
df_ext.loc[df_ext['Animal'] == '2021-05-21 LC GcAMP Tone Only Extinction_3.csv', 'Animal'] = 206
df_ext.loc[df_ext['Animal'] == '2021-05-21 LC GcAMP Tone Only Extinction_4.csv', 'Animal'] = 209
```

```python
# apply 6 Hz low-pass filter
df_train_lowfilt = fp_dat.lowpass_filter(df_train, x1='465nm', x2='405nm', 
                                          cutoff=6, fsamp=100, order=4)
df_tone_lowfilt = fp_dat.lowpass_filter(df_tone, x1='465nm', x2='405nm', 
                                          cutoff=6, fsamp=100, order=4)
df_ext_lowfilt = fp_dat.lowpass_filter(df_ext, x1='465nm', x2='405nm', 
                                          cutoff=6, fsamp=100, order=4)

# resample to 10 Hz
df_train_resamp = fp_dat.resample_data(df_train_lowfilt, freq=10)
df_tone_resamp = fp_dat.resample_data(df_tone_lowfilt, freq=10)
df_ext_resamp = fp_dat.resample_data(df_ext_lowfilt, freq=10)


# apply linear fit for motion correction to obtain dFF
df_train_fit = fp_dat.fit_linear(df_train_resamp, Y_sig='465nm', Y_ref='405nm')
df_tone_fit = fp_dat.fit_linear(df_tone_resamp, Y_sig='465nm', Y_ref='405nm')
df_ext_fit = fp_dat.fit_linear(df_ext_resamp, Y_sig='465nm', Y_ref='405nm')

# trial level data
df_tone_train_trials = tfc_dat.trials_df(df_train_fit, yvar='465nm_dFF', session='train')
df_tone_tone_trials = tfc_dat.trials_df(df_tone_fit, yvar='465nm_dFF',us_dur = 0, session='tone')
df_tone_ext_trials = tfc_dat.trials_df(df_ext_fit, yvar='465nm_dFF', session='extinction', iti_dur=70,us_dur=0)

#drop animals with bad or no signal
df_tone_train_trials = df_tone_train_trials.query("~(Animal == 207)").reset_index(drop=True)

```

```python
# label the days and concatenate all days into 1 dataframe for the tone alone group
df_tone_train_trials['Day'] = 1
df_tone_tone_trials['Day'] = 2
df_tone_ext_trials['Day'] = 3

df_tone_only = pd.concat([df_tone_train_trials, df_tone_tone_trials, df_tone_ext_trials]).reset_index()
```

# Looking at Tone Responses by Day and Group

```python
# Get the peak dFF to the tone onset in the tone only group on each day

tone_train_prepost = fp_dat.calc_pre_post(df_tone_train_trials, event = 'tone', t_pre=(-2,0), t_post=(0,2), measure = 'max')
tone_tone_prepost = fp_dat.calc_pre_post(df_tone_tone_trials, event = 'tone',  t_pre=(-2,0), t_post=(0,2), measure = 'max')
tone_ext_prepost = fp_dat.calc_pre_post(df_tone_ext_trials, event = 'tone',  t_pre=(-2,0), t_post=(0,2), measure = 'max')
```

```python
# Get the peak dFF to the tone onset in the TFC group on each day

tfc_train_prepost = fp_dat.calc_pre_post(df_tfc_train_trials, event = 'tone',  t_pre=(-2,0), t_post=(0,2), measure = 'max')
tfc_tone_prepost = fp_dat.calc_pre_post(df_tfc_tone_trials, event = 'tone',  t_pre=(-2,0), t_post=(0,2), measure = 'max')
tfc_ext_prepost = fp_dat.calc_pre_post(df_tfc_ext_trials, event = 'tone',  t_pre=(-2,0), t_post=(0,2), measure = 'max')
```

```python
# Plot pre and post tone responses in the two groups on Day 1

fig, ax = plt.subplots(1, 2, figsize = (15,10))
fig.suptitle('Day 1', y = 1.05,fontsize = 40)

sns.pointplot(x = 'epoch', y = '465nm_dFF_norm', hue = 'Animal', order = ['pre-tone', 'post-tone'], ax = ax[0], data = tone_train_prepost)
plt.tight_layout()
ax[0].legend('')
ax[0].title.set_text('Tone Only')
ax[0].set_ylabel('Peak dFF')
ax[0].set_xlabel('')
ax[0].set_ylim(-0.5,8)

sns.pointplot(x = 'epoch', y = '465nm_dFF_norm', hue = 'Animal', order = ['pre-tone', 'post-tone'], ax = ax[1], data = tfc_train_prepost)
plt.tight_layout()
ax[1].legend('')
ax[1].title.set_text('TFC')
ax[1].set_ylabel('Peak dFF')
ax[1].set_xlabel('')
ax[1].set_ylim(-0.5,8)

plt.tight_layout()

```

```python
# Plot pre and post tone responses in the two groups on Day 2

fig, ax = plt.subplots(1, 2, figsize = (15,10))
fig.suptitle('Day 2', y = 1.05,fontsize = 40)

sns.pointplot(x = 'epoch', y = '465nm_dFF_norm', hue = 'Animal', order = ['pre-tone', 'post-tone'], ax = ax[0], data = tone_tone_prepost)
plt.tight_layout()
ax[0].legend('')
ax[0].title.set_text('Tone Only')
ax[0].set_ylabel('Peak dFF')
ax[0].set_xlabel('')
ax[0].set_ylim(0,6)

sns.pointplot(x = 'epoch', y = '465nm_dFF_norm', hue = 'Animal', order = ['pre-tone', 'post-tone'], ax = ax[1], data = tfc_tone_prepost)
plt.tight_layout()
ax[1].legend('')
ax[1].title.set_text('TFC')
ax[1].set_ylabel('Peak dFF')
ax[1].set_xlabel('')
ax[1].set_ylim(0,6)

plt.tight_layout()
```

```python
# Plot pre and post tone responses in the two groups on Day 3

fig, ax = plt.subplots(1, 2, figsize = (15,10))
fig.suptitle('Day 3', y = 1.05,fontsize = 40)

sns.pointplot(x = 'epoch', y = '465nm_dFF_norm', hue = 'Animal', order = ['pre-tone', 'post-tone'], ax = ax[0], data = tone_ext_prepost)
plt.tight_layout()
ax[0].legend('')
ax[0].title.set_text('Tone Only')
ax[0].set_ylabel('Peak dFF')
ax[0].set_xlabel('')
ax[0].set_ylim(-0.5,3)

sns.pointplot(x = 'epoch', y = '465nm_dFF_norm', hue = 'Animal', order = ['pre-tone', 'post-tone'], ax = ax[1], data = tfc_ext_prepost)
plt.tight_layout()
ax[1].legend('')
ax[1].title.set_text('TFC')
ax[1].set_ylabel('Peak dFF')
ax[1].set_xlabel('')
ax[1].set_ylim(-0.5,4)
plt.tight_layout()
```

```python
# combine all days data from tone only into one dataframe
tone_only_prepost = pd.concat([tone_train_prepost,tone_tone_prepost,tone_ext_prepost]).reset_index()
```

```python
# combine all days data from TFC into one dataframe
tfc_prepost = pd.concat([tfc_train_prepost,tfc_tone_prepost,tfc_ext_prepost]).reset_index()
```

```python
# add group label to prepost data frame
tone_only_prepost['Group'] = 'Tone Only'
tfc_prepost['Group'] = 'TFC'
```

```python
# combine tone only and TFC data into one dataframe
all_prepost = pd.concat([tone_only_prepost, tfc_prepost]).reset_index()
```

```python
# Plot peak dFF of tone only and TFC groups on each day
pal = ['dodgerblue', '#9147B1']
fig, ax = plt.subplots(figsize = (12,10))

p = sns.pointplot(x = 'Day', y = '465nm_dFF_norm', hue = 'Group', 
                ci = 68, hue_order=['TFC','Tone Only'], palette = pal, capsize=0.1, 
                data = all_prepost.query('epoch == "post-tone"'),  errwidth = 6, scale = 2)

ax.set_ylabel(r'Peak $\Delta$F/F', fontsize = 60)
ax.set_xlabel('Day', fontsize = 60)
ax.set_xticklabels(['1','2','3'], fontsize = 50)
plt.yticks(fontsize = 50)
plt.setp(ax.collections, sizes = [800])
p.legend_.set_title(None)
leg = plt.legend(bbox_to_anchor=(1.3, 1), loc = 'upper right',fontsize = 40)
leg.get_texts()[0].set_text('TFC (Extinction)')
leg.get_texts()[1].set_text('Tone Only (Habituation)')
#ax.set_ylim(0,2)

#plt.tight_layout()
plt.savefig('/Users/jake/Box/WiltgenLab expts/JW LC/LC GcAMP Combined Analyses/TFC vs Tone Only Plot')
```
